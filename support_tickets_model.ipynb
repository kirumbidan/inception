*****USE instruction and category to predict intent*****

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
features = ['instruction', 'category']
X = df[features]
y = df['intent']
```

```python
X.isna().value_counts()
df[features].isna().value_counts()
df['intent'].unique()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y,
                                                      train_size=0.8,
                                                      random_state=0)
```

```python
X_train.head()
```

Model built using a RandomForestClassifier

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), features)],
    remainder='passthrough'
)

pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(n_estimators=100, oob_score=True, random_state=0))
])
pipe.fit(X_train, y_train)
print("OOB Score: " + str(pipe.named_steps['model'].oob_score_))
```

```python
from sklearn.metrics import classification_report, accuracy_score
preds = pipe.predict(X_valid)
print(f"Accuracy Score: {accuracy_score(y_valid, preds)}")
print(classification_report(y_valid, preds))
```

***Use tfidfvectorizer***

***OneHotEncoding Categorical Columns and Training with a RandomForestClassifier has 50% accuracy model***

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_preprocessor = ColumnTransformer(
    transformers=[
        ('subject_tfidf', TfidfVectorizer(stop_words='english', max_features=500), 'category'),
        ('desc_tfidf', TfidfVectorizer(stop_words='english', max_features=1000), 'instruction')
    ],
    remainder='passthrough'
)
```

```python
pipe_2 = Pipeline(steps=[
    ('preprocessor', tfidf_preprocessor),
    ('model', RandomForestClassifier(n_estimators=500, oob_score=True, random_state=0))
])

pipe_2.fit(X_train, y_train)

preds_2 = pipe_2.predict(X_valid)
print("OOB Score with TfidfVectorizer: " + str(pipe_2.named_steps['model'].oob_score_))
```

```python
print(f"Accuracy Score: {accuracy_score(y_valid, preds_2)}")
print(classification_report(y_valid, preds_2))
```

***Vectorizing Columns with TfidfVectorizer makes the model have 98% accuracy on training and validation sets!***

Try Running a Gridsearchcv to find the best parameter combinations and see if the model can be more accurate

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'model__n_estimators': [200, 500],
    'model__max_depth': [20, 50, None],
    'model__min_samples_split': [2, 5, 10],
    'model__max_features': ['sqrt', 'log2']
}
grid_search = GridSearchCV(pipe_2, param_grid, cv=3, scoring='accuracy', verbose=1)
grid_search.fit(X_train, y_train)

print(f"Best Score: {grid_search.best_score_}")
print(f"Best Params: {grid_search.best_params_}")
```

```python
grid_preds = grid_search.predict(X_valid)
print(f"Accuracy Score: {accuracy_score(y_valid, grid_preds)}")
print(classification_report(y_valid, grid_preds))
```

***The model is 98% accurate after finding the best parameters with a GridSearch
